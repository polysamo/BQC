```python
from quantumnet.components import Network, Host
from quantumnet.components import ServerNode, ClientNode, RegularNode
from quantumnet.objects import Qubit, Logger
import random
```

```python
# instanciando a rede
rede = Network()

# instanciando hosts (com tipos específicos, se necessário)
h1 = ServerNode(1)
h2 = ClientNode(2)
h3 = ClientNode(3)


# criando as conexões
h1.add_connection(2)
h1.add_connection(3)
h2.add_connection(3)

# adicionando hosts à rede
rede.add_host(h1)
rede.add_host(h2)
rede.add_host(h3)

# plotando a rede
rede.draw()
```

```python
# o acesso dos nós e arestas do grafo da rede pode ser feito diretamente pela rede
print(rede.nodes)
print(rede.edges)
# ou pela propriedade graph da rede
print(rede.graph.nodes)
print(rede.graph.edges)
```

```python
rede = Network()
rede.set_ready_topology('grade',4,3,3)  # Exemplo de uma grade 3x3 com 2 clientes e 1 servidor
rede.draw()
Logger.activate(Logger)
```
